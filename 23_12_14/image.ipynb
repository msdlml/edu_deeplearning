```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open('justice.jpg')
# plt.imshow(img)
# plt.show()

pixel = np.array(img)
print(pixel.shape)
plt.imshow(pixel)
plt.show()
#crop(, , , )
crop_img = img.crop((30,100,150,300))
plt.imshow(crop_img)
# plt.show()

##   ,      
resize_img = img.resize((100,100))
# plt.imshow(resize_img)
# plt.show()

rotate_img = img.rotate(180)
#plt.imshow(rotate_img)
plt.show()
```

```python
color_img = Image.open('fruits.jpg')

color_pixel = np.array(color_img)
#    
plt.imshow(color_pixel)
plt.show()

# 
# pixel rgb   
print(color_pixel.shape)
gray_pixel = color_pixel.copy()

for y in range(gray_pixel.shape[0]):
    for x in range(gray_pixel.shape[1]):
        gray_pixel[y,x] = int(np.mean(gray_pixel[y,x]))

plt.imshow(gray_pixel)
plt.show()

# 3  >> 2 
gray2d_pixel = gray_pixel[:,:,0]

plt.imshow(gray2d_pixel, cmap='gray')
plt.show()
```
